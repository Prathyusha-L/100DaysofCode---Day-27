```python
import pandas as pd
from matplotlib import pyplot
from matplotlib import pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing,Holt
from pylab import rcParams
rcParams['figure.figsize']=20,5
```

```python
series=pd.read_csv('C:/Users/Prathyu Lachireddy/Desktop/BP/Electric_Production.csv',header=0, index_col=0)
series
```

```python
data=series[1:50]

fig_dims = (7, 4)
fig, ax = plt.subplots(figsize=fig_dims)
fit1=SimpleExpSmoothing(data).fit(smoothing_level=0.2,optimized=False)

fit2=SimpleExpSmoothing(data).fit(smoothing_level=0.8,optimized=False)
```

```python
plt.figure(figsize=(18,9))
plt.plot(series[1:50],marker='o',color='black')
plt.xticks(rotation=30)

plt.plot(fit1.fittedvalues,marker='o',color='blue')
plt.plot(fit2.fittedvalues,marker='o',color='red')
```

```python
fit1 = Holt(data).fit()

fit2 = Holt(data,exponential = False).fit()

fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)
plt.plot(data,marker = 'o',color = 'black')
plt.xticks(rotation = 30)

plt.plot(fit2.fittedvalues,marker = 'o',color = 'red')
```
